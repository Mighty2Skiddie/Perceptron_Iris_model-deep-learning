```python
X=data.drop('Class',axis=1)
Y=data['Class']
```

```python
from sklearn.metrics import accuracy_score
```

```python
class Perceptron:

  def __init__(self):
      self.w = None
      self.b = None

  def model(self, x):
      return 1 if (np.dot(self.w, x) >= self.b) else 0

  def predict(self, X):
      Y = []
      for x in X:
        result = self.model(x)
        Y.append(result)
      return np.array(Y)

  def fit(self, X, Y, epochs = 1):
      self.w = np.ones(X.shape[1])
      self.b = 0

      accuracy = {}

      for i in range(epochs):
          for x, y in zip(X, Y):
            y_pred = self.model(x)
            if y == 1 and y_pred == 0:
              self.w = self.w + x
              self.b = self.b + 1
            elif y == 0 and y_pred == 1:
              self.w = self.w - x
              self.b = self.b - 1
          accuracy[i] = accuracy_score(self.predict(X), Y)

      plt.plot(accuracy.values())
      plt.xlabel('Epochs')
      plt.ylabel('Accuracy')
      plt.show()

```

```python
perceptron=Perceptron()
```
